```python
from transformers import RobertaModel, RobertaTokenizer, RobertaForSequenceClassification
```

```python
import torch
```

```python
import pandas as pd
```

```python
model_name = 'cointegrated/roberta-large-cola-krishna2020'
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
tokenizer = RobertaTokenizer.from_pretrained(model_name)
```

```python
model = RobertaForSequenceClassification.from_pretrained(model_name).to(torch_device)
```

```python
fluency = []
```

```python
path = "data/Testing/pseudo_parallel/pseudo generated and actual text.csv"
df = pd.read_csv(path)
```

```python
for _, row in df.iterrows():
  sentence = list(row)[1]
  inputs = tokenizer(sentence, return_tensors="pt")
  with torch.no_grad():
      logits = model(**inputs).logits
  predicted_class_id = logits.argmax().item()
  fluency.append(predicted_class_id)
```

```python
fluency
```

```python
fluency_df = pd.DataFrame(fluency)
fluency_df.to_csv("data/Testing/pseudo_parallel/pseudo_fluency.csv")
```
