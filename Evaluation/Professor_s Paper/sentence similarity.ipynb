```python
# import required libraries
import pandas as pd
import numpy as np
from numpy.linalg import norm
from sentence_transformers import SentenceTransformer
```

```python
path = "data/Testing/pseudo_parallel/pseudo generated and actual text.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
similarities = []
```

```python
model = SentenceTransformer('sentence-transformers/stsb-roberta-base-v2')
for _, row in df.iterrows():
  sentences = list(row)
  embeddings = model.encode(sentences)
 
  # define two lists or array
  A = np.array(embeddings[0])
  B = np.array(embeddings[1])

  # compute cosine similarity
  cosine = np.dot(A,B)/(norm(A)*norm(B))
  print("Cosine Similarity:", cosine)
  similarities.append(cosine)
```

```python
similarities
```

```python
sim_df = pd.DataFrame(similarities)
sim_df.to_csv("data/Testing/pseudo_parallel/pseudo_similarity_score.csv")
```
